```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python


# Define the linear function
def linear_function(x):
    return 2 * x + 3

# Generate x values
x_values = np.linspace(-10, 10, 400)  # Generate 400 points between -10 and 10

# Calculate corresponding y values using the linear function
y_values = linear_function(x_values)

# Plot the linear function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='f(x) = 2x + 3', color='blue')
plt.title('Graph of Linear Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def quadratic_function(x):
    return x**2 - 4*x + 3

# Generate x values
x_values = np.linspace(-2, 6, 400)  # Generate 400 points between -2 and 6

# Calculate corresponding y values using the quadratic function
y_values = quadratic_function(x_values)

# Plot the quadratic function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = x^2 - 4x + 3$', color='red')
plt.title('Graph of Quadratic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def exponential_function(x):
    return np.exp(x)

# Generate x values
x_values = np.linspace(-5, 5, 400)  # Generate 400 points between -5 and 5

# Calculate corresponding y values using the exponential function
y_values = exponential_function(x_values)

# Plot the exponential function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = e^x$', color='green')
plt.title('Graph of Exponential Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def logarithmic_function(x):
    return np.log10(x)

def ln_function(x):
    return np.log(x)

# Generate x values (excluding 0)
x_values = np.linspace(0.1, 5, 400)  # Generate 400 points between 0.1 and 5

# Calculate corresponding y values using the logarithmic function
y_values = logarithmic_function(x_values)
y2_values = ln_function(x_values)
# Plot the logarithmic function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = log_{10}(x)$', color='purple')
plt.plot(x_values, y2_values, label='$f(x) = ln(x)$', color='yellow')
plt.title('Graph of Logarithmic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x_values = np.linspace(-2*np.pi, 2*np.pi, 1000)  # Generate 1000 points from -2*pi to 2*pi

# Calculate trig function values
sine_values = np.sin(x_values)
cosine_values = np.cos(x_values)
tangent_values = np.tan(x_values)

# Plot the trig functions
plt.figure(figsize=(10, 6))
plt.plot(x_values, sine_values, label='sin(x)', color='blue')
plt.plot(x_values, cosine_values, label='cos(x)', color='red')
plt.plot(x_values, tangent_values, label='tan(x)', color='green')
plt.axhline(0, color='black', linewidth=0.5)  # Add x-axis line
plt.axvline(0, color='black', linewidth=0.5)  # Add y-axis line
plt.title('Common Trigonometric Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-3, 3)  # Set y-axis limits for better visualization
plt.legend()
plt.grid(True)
plt.show()
```

```python
def power_function(x):
    return x**3

# Generate x values
x_values = np.linspace(-5, 5, 400)  # Generate 400 points between -5 and 5

# Calculate corresponding y values using the power function
y_values = power_function(x_values)

# Plot the power function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = x^3$', color='cyan')
plt.title('Graph of Power Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def gaussian_function(x, mu, sigma):
    return np.exp(-((x - mu)**2) / (2 * sigma**2)) / (sigma * np.sqrt(2 * np.pi))

# Generate x values
x_values = np.linspace(-5, 5, 400)  # Generate 400 points between -5 and 5

# Calculate corresponding y values using the Gaussian function
mu = 0  # Mean
sigma = 1  # Standard deviation
y_values = gaussian_function(x_values, mu, sigma)

# Plot the Gaussian function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = e^{-x^2}$', color='magenta')
plt.title('Graph of Gaussian Function (Normal Distribution)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def f(x):
    return x**2

# Define the integration limits
a = 0
b = 2

# Calculate the integral and its value
integral_value, _ = quad(f, 0, 3)

# Generate x values for plotting
x = np.linspace(a-2, b+2, 100)
y = f(x)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the function curve
plt.plot(x, y, label='f(x) = x^2')

# Fill the shaded area under the curve
plt.fill_between(x, y, alpha=0.3, label='Shaded Area',
                 where = (x>0) & (x <=3.5))

# Annotate the integral value
plt.text(1.2, 10.5, f'$\int_0^3 f(x)dx$ = {integral_value:.2f}', fontsize=12, color='black')

# Set labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Integral with Shaded Area')

# Add legend
plt.legend()

# Show the plot
plt.grid()
plt.show()
```
